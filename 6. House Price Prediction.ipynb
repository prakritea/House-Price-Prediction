```python
features_with_na = [features for features in dataset.columns if (dataset[features].isnull().sum() > 1)]

for features in features_with_na:
    print(features,np.round(dataset[features].isnull().mean(),4),"%missing values")

```

```python
for features in features_with_na:
    data = dataset.copy()
    data[features] = np.where(data[features].isnull(),1,0)
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.title(features)
    plt.show()
```

```python
#numerical values are of two types... continuous and discrete
```

```python
discrete_features = [ features for features in dataset.columns if (len(dataset[features].unique()) < 25) and (features not in year_features + ['Id'])] 
discrete_features
```

```python
#relation between discrete features and SalePrice
for features in discrete_features:
    data = dataset.copy()
    data.groupby(features)['SalePrice'].median().plot.bar(color = 'pink')
    plt.xlabel(features)
    plt.ylabel('SalePrice')
    plt.title(features)
    plt.show()
```

```python
#continous variable
```

```python
continous_data = [features for features in numerical_features if features not in discrete_features + year_features+['Id']]
continous_data
```

```python
len(continous_data)
```

```python
for features in continous_data:
    data = dataset.copy()
    data[features].hist(bins=25)
    plt.xlabel(features)
    plt.title(features)
    plt.show()
```

```python
for features in continous_data:
    data = dataset.copy()
    if 0 in data[features].unique():
        pass
    else:
        data[features] = np.log(data[features])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[features],data['SalePrice'])
        plt.xlabel(features)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
#outliers
for features in continous_data:
    data = dataset.copy()
    if 0 in data[features].unique():
        pass
    else:
        data[features] = np.log(data[features])
        data.boxplot(column = features)
        plt.ylabel(features)
        plt.title(features)
        plt.show()
```

# Feature Engineering

```python
# filling missing values
```

```python
dataset.isnull().sum()
```

```python
dataset['LotFrontage'] = dataset['LotFrontage'].fillna(dataset['LotFrontage'].mean())
```

```python
dataset.drop(['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1','BsmtFinType2', 'Electrical', 'PoolQC', 'Fence', 'MiscFeature'], axis = 1, inplace = True)
```

```python
#filling missing values on categorical data
dataset['MasVnrType'].fillna(dataset['MasVnrType'].mode()[0], inplace=True)
dataset['FireplaceQu'].fillna('None', inplace=True)
dataset['GarageType'].fillna('None', inplace=True)
dataset['GarageFinish'].fillna('None', inplace=True)
dataset['GarageQual'].fillna('None', inplace=True)
dataset['GarageCond'].fillna('None', inplace=True)
```

```python
#handling numerical values

dataset['MasVnrArea'].fillna(dataset['MasVnrArea'].median(), inplace=True)
dataset['GarageYrBlt'].fillna(dataset['GarageYrBlt'].median(), inplace=True)
```
